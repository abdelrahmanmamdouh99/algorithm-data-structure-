```python
import matplotlib.pyplot as plt
import numpy as np
import math

x=[10**i for i in range(10)]#logrithmic scale

funcs=[lambda x: 8*x,
     lambda x: 4*x*math.log(x),
     lambda x:2*x**2,
     lambda x:x**3,
     #lambda x:2**x
      ]

ys=[]

for func in funcs:
    ys.append(list(map(func, x)))#the values of y(x) axis as list to each func (list of lists)

print(ys)
for y in ys:
    plt.plot(x,y)
plt.yscale('log')
plt.xscale('log')
```

```python
# as proof of R-3.9

x = [x/100 for x in range(1, 500)]
y1 = list(map(lambda x: 4*math.log(x,2), x))
y2 = x

y_1=list(map(lambda x:2*x ,y1))
y_2=list(map(lambda x:2*x, y2))

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.show()
# ad(n) -> O(f(n))
```
